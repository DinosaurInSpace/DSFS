```python
from matplotlib import pyplot as plt
```

```python
# Line graph

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.plot(x, y, color ='blue', marker = 'x', linestyle='solid')

plt.title("Such data")
plt.ylabel("Such y")
plt.xlabel("Such x")
plt.axis([0, 10, 0, 10])

plt.show
```

```python
# Bar graph

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.bar(x, y)

plt.title("Such data")
plt.ylabel("Such y")
plt.xlabel("Such x")
plt.axis([0, 10, 0, 10])

plt.show
```

```python
# Scatter plot

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.scatter(x, y)

plt.title("Such data")
plt.ylabel("Such y")
plt.xlabel("Such x")
plt.axis([0, 10, 0, 10])

plt.show
```
